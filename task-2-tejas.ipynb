```python
import numpy as np
import numpy.fft as fft
import matplotlib.pyplot as plt
```

```python
n = 512
x = np.linspace(0, 2*np.pi, n)
f = np.sin(x)
```

```python
plt.plot(x, f)
plt.show()
```

```python
#Fourier Transform of function(x)
F = fft.fft(f)
```

```python
w = fft.fftfreq(n)
plt.plot(w, F)
plt.show()
```

```python
plt.plot(w, np.abs(F))
plt.show()
```

```python
plt.plot(w, np.log(np.abs(F)))
plt.show()
```

```python
offset = 1.
amplitude = 2.
frequency = 16.
phase = np.pi
f = offset + amplitude * np.sin(frequency*x + phase)
F = fft.fft(f)
w = fft.fftfreq(n)
```

```python
plt.plot(w,np.abs(F))
```
